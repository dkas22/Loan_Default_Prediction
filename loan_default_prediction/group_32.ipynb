# Imputing Missing values

```python
new_data.describe()
```

```python
# took mean beacause there in no much difference in mean and median
new_data['Income'].fillna(new_data['Income'].mean(), inplace=True)
new_data['Score1'].fillna(new_data['Score1'].mean(), inplace=True)
new_data['Score2'].fillna(new_data['Score2'].mean(), inplace=True)
new_data['Score3'].fillna(new_data['Score3'].mean(), inplace=True)
new_data['Score4'].fillna(new_data['Score4'].mean(), inplace=True)
new_data['Score5'].fillna(new_data['Score5'].mean(), inplace=True)
```

```python
new_data['Loan type'].fillna(new_data['Loan type'].value_counts().index[0],inplace = True)
new_data['Occupation type'].fillna(new_data['Occupation type'].value_counts().index[0],inplace = True)
new_data['Age'].fillna(new_data['Age'].value_counts().index[0],inplace = True)
```

```python
X = new_data.iloc[:,:-1]
```

```python
Y = new_data.iloc[:,-1:]
```

```python
from sklearn.model_selection import cross_val_score
```

# Data scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler().fit(X)
```

```python
scaler
```

```python
scaled_x = scaler.transform(X)
```

```python
scaled_x.mean(axis=0)
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
for i in range(5):
    LogReg = LogisticRegression()
    print("Accuracy Using Logistic Regression: ",cross_val_score(LogReg,scaled_x,Y.iloc[:,0],cv=10, scoring='accuracy').mean()*100)
    print("Precision Using Logistic Regression: ",cross_val_score(LogReg,scaled_x,Y.iloc[:,0],cv=10,scoring='precision').mean()*100)
    print("F1 score Using Logistic Regression: ",cross_val_score(LogReg,scaled_x,Y.iloc[:,0],cv=10,scoring='f1').mean()*100)
    print("Recall Using Logistic Regression: ",cross_val_score(LogReg,scaled_x,Y.iloc[:,0],cv=10,scoring='recall').mean()*100)
    print()
```

# K-Nearest Neighbors Classifier

```python
#  Determing best value of k using cross_validation technique
from sklearn.neighbors import KNeighborsClassifier
Accuracy = []
for i in range(1,41):
    knnclassifier = KNeighborsClassifier(n_neighbors=i)
    Accuracy.append(cross_val_score(knnclassifier,scaled_x,Y.iloc[:,0],cv=10, scoring='accuracy').mean())
```

```python
plt.figure(figsize=(15,8))
plt.plot(range(1, 41), Accuracy, color='blue', linestyle='dashed', marker='o',
         markerfacecolor='blue', markersize=10)
plt.title('Accuracy vs K value')
plt.xlabel('K value')
plt.ylabel('Cross validation Score Mean')
```

```python
# from cross validation we are getting K=3 as best value
for i in range(5):
    knnclassifier = KNeighborsClassifier(n_neighbors=3)
    print("Accuracy using KNN classifier: ",cross_val_score(knnclassifier,scaled_x,Y.iloc[:,0],cv=10, scoring='accuracy').mean())
    print("Precision using KNN classifier: ",cross_val_score(knnclassifier,scaled_x,Y.iloc[:,0],cv=10, scoring='precision').mean())
    print("F1 Score using KNN classifier: ",cross_val_score(knnclassifier,scaled_x,Y.iloc[:,0],cv=10, scoring='f1').mean())
    print("recall using KNN classifier: ",cross_val_score(knnclassifier,scaled_x,Y.iloc[:,0],cv=10, scoring='recall').mean())
    print()
    
```

# Decision Tree Classifier

```python
from sklearn import tree
for i in range(5):
    dtc = tree.DecisionTreeClassifier()
    print("Accuracy Using Decision Tree: ",cross_val_score(dtc,scaled_x,Y.iloc[:,0],cv=10, scoring='accuracy').mean()*100)
    print("Precision Using Decision Tree: ",cross_val_score(dtc,scaled_x,Y.iloc[:,0],cv=10,scoring='precision').mean()*100)
    print("F1 score Using Decision Tree: ",cross_val_score(dtc,scaled_x,Y.iloc[:,0],cv=10,scoring='f1').mean()*100)
    print("Recall Using Decision Tree: ",cross_val_score(dtc,scaled_x,Y.iloc[:,0],cv=10,scoring='recall').mean()*100)
    print()
```

# Random Forest Classifier

```python
# accuracy vs max_depth
from sklearn.ensemble import RandomForestClassifier
for i in range(1,10):
    RF = RandomForestClassifier(n_estimators=100, max_depth=i, random_state=1)
    print(cross_val_score(RF,scaled_x,Y.iloc[:,0],cv=10, scoring='accuracy').mean()*100)
```

```python
RF = RandomForestClassifier(n_estimators=100, max_depth=10, random_state=1)
print("Accuracy Using Random Forest: ",cross_val_score(RF,scaled_x,Y.iloc[:,0],cv=10, scoring='accuracy').mean()*100)
print("Precision Random Forest: ",cross_val_score(RF,scaled_x,Y.iloc[:,0],cv=10,scoring='precision').mean()*100)
print("F1 score Using Random Forest: ",cross_val_score(RF,scaled_x,Y.iloc[:,0],cv=10,scoring='f1').mean()*100)
print("Recall Using Random Forest: ",cross_val_score(RF,scaled_x,Y.iloc[:,0],cv=10,scoring='recall').mean()*100)
```

# Gradient Boosting Classifier or XGBoost

```python
from sklearn.ensemble import GradientBoostingClassifier
clf = GradientBoostingClassifier(n_estimators=100, learning_rate=1.0,max_depth=5, random_state=0)
print("Accuracy Using XGBoost: ",cross_val_score(clf,scaled_x,Y.iloc[:,0],cv=10, scoring='accuracy').mean()*100)
print("Precision Using XGBoost: ",cross_val_score(clf,scaled_x,Y.iloc[:,0],cv=10,scoring='precision').mean()*100)
print("F1 score Using XGBoost: ",cross_val_score(clf,scaled_x,Y.iloc[:,0],cv=10,scoring='f1').mean()*100)
print("Recall Using XGBoost: ",cross_val_score(clf,scaled_x,Y.iloc[:,0],cv=10,scoring='recall').mean()*100)
```
